# Algunos ejemplos de gráficas

```python
import pandas as pd
import datetime

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["font.family"] = "segoe ui"

import numpy as np
```

## 1. Gráficas de líneas (serie de tiempo)
Se nos solicita realizar un gráfico del *PIB* nominal de Guatemala. Dicho gráfico debería tener:
- Título con el nombre de la serie graficada.
- Subtítulo con el periodo graficado.
- La línea debe contar también con puntos.
- Fuente de la información.

```python
# Cargamos los datos (esto se verá con más detalle en la sesión de pandas)

pib = pd.read_csv('pib_nominal.csv')
pib.fecha = pd.to_datetime(pib.fecha, format='%d/%m/%Y')
```

```python
# Creamos la figura y sus respectivos ejes. 

fig, axis1 = plt.subplots()
axis1.plot(pib.fecha, pib.y, **{'marker': 'o'}) # Al objeto axis1 se le aplica un método para dibujar en él la serie.



plt.show()
```

```python
# Agregamos títulos a los ejes al gráfico

fig, axis1 = plt.subplots()
axis1.plot(pib.fecha, pib.y,  **{'marker': 'o'}) # Al objeto axis1 se le aplica un método para dibujar en él la serie.

axis1.set_xlabel(
    'Fecha', 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)
axis1.set_ylabel(
    'Millones de Quetzales' , 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)

plt.show()
```

```python
# Agregamos título y subtítulo a la gráfica (opcion 1: separando el título utilizando \n)

fig, axis1 = plt.subplots()
axis1.plot(pib.fecha, pib.y,  **{'marker': 'o'}) # Al objeto axis1 se le aplica un método para dibujar en él la serie.

axis1.set_xlabel(
    'Fecha', 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)
axis1.set_ylabel(
    'Millones de Quetzales' , 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)

axis1.set_title(
    'Producto Interno Bruto Nominal \n 2013Q1 a 2019Q4', 
    **{'fontsize':'x-large'}
)

plt.show()
```

```python
# Agregamos título y subtítulo a la gráfica (opcion 2: utilizando el rederizado de latex)

fig, axis1 = plt.subplots()
axis1.plot(pib.fecha, pib.y,  **{'marker': 'o'}) # Al objeto axis1 se le aplica un método para dibujar en él la serie.

axis1.set_xlabel(
    'Fecha', 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)
axis1.set_ylabel(
    'Millones de Quetzales' , 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)



axis1.set_title(
    r'$\bf{Producto \; Interno \; Bruto \; de \; Guatemala}$' + '\n 2013Q1 a 2019Q4',
    **{'fontname':'segoe ui','fontsize':'x-large'}
)

plt.show()
```

```python
# Agregamos título y subtítulo a la gráfica (opcion 3: utilizando suptitle)

fig, axis1 = plt.subplots()
axis1.plot(pib.fecha, pib.y,  **{'marker': 'o'}) # Al objeto axis1 se le aplica un método para dibujar en él la serie.

axis1.set_xlabel(
    'Fecha', 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)
axis1.set_ylabel(
    'Millones de Quetzales' , 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)


plt.suptitle(
    'Producto Interno Bruto de Guatemala',
    **{'fontname':'segoe ui','fontsize':'x-large', 'fontweight': 'bold'}
)
axis1.set_title(
    '2013Q1 a 2019Q4',
    **{'fontname':'segoe ui','fontsize':'large'}
)

plt.show()
```

```python
# Agregamos la fuente

fig, axis1 = plt.subplots()
axis1.plot(pib.fecha, pib.y,  **{'marker': 'o'}) # Al objeto axis1 se le aplica un método para dibujar en él la serie.

axis1.set_xlabel(
    'Fecha', 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)
axis1.set_ylabel(
    'Millones de Quetzales' , 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)



axis1.set_title(
    r'$\bf{Producto \; Interno \; Bruto \; de \; Guatemala}$' + '\n 2013Q1 a 2019Q4',
    **{'fontname':'segoe ui','fontsize':'x-large'}
)

txt="Fuente: Banco de Guatemala"
plt.figtext(
    0.5, -0.1, txt,
    **{'fontname':'segoe ui','fontsize':'large','wrap':True, 'horizontalalignment': 'center'}
)

plt.show()
```

```python
# Guardamos la imagen

fig, axis1 = plt.subplots()
axis1.plot(pib.fecha, pib.y,  **{'marker': 'o'}) # Al objeto axis1 se le aplica un método para dibujar en él la serie.

axis1.set_xlabel(
    'Fecha', 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)
axis1.set_ylabel(
    'Millones de Quetzales' , 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)



axis1.set_title(
    r'$\bf{Producto \; Interno \; Bruto \; de \; Guatemala}$' + '\n 2013Q1 a 2019Q4',
    **{'fontname':'segoe ui','fontsize':'x-large'}
)

txt="Fuente: Banco de Guatemala"
plt.figtext(
    0.5, -0.1, txt,
    **{'fontname':'segoe ui','fontsize':'large','wrap':True, 'horizontalalignment': 'center'}
)

plt.savefig('pib1.png')

plt.show()
```

## 2. Dos series en el mismo eje
Se solicuta que a la figura anterior se le añada el promedio móvil de 4 trimestres.

```python
pib['sm_y'] = pib.y.rolling(4).mean()
```

```python
fig, axis1 = plt.subplots()

axis1.plot(pib.fecha, pib.y,  **{'marker': 'o', 'label': 'Nivel'}) # Al objeto axis1 se le aplica un método para dibujar en él la serie.
axis1.plot(pib.fecha, pib.sm_y,  **{'marker': 'o', 'label': 'Promedio móvil*'})

axis1.legend()

axis1.set_xlabel(
    'Fecha', 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)
axis1.set_ylabel(
    'Millones de Quetzales' , 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)



axis1.set_title(
    r'$\bf{Producto \; Interno \; Bruto \; de \; Guatemala}$' + '\n 2013Q1 a 2019Q4',
    **{'fontname':'segoe ui','fontsize':'x-large'}
)

txt="Fuente: Banco de Guatemala \n *Promedio móvil de 4 trimestres"
plt.figtext(
    0.5, -0.1, txt,
    **{'fontname':'segoe ui','fontsize':'medium','wrap':True, 'horizontalalignment': 'center'}
)

plt.savefig('pib2.png')

plt.show()
```

## 3. Dos ejes en el mismo gráfico
Se solicita que grafique las variaciones interanuales junto al gráfico anterior.

```python
pib['v_y'] = pib.y.pct_change(periods=4) 
pib['v_sm_y'] = pib.sm_y.pct_change(periods=4) 
```

```python
fig, axis = plt.subplots(2,1, figsize=(12,12))

# Ejes superiores

axis[0].plot(pib.fecha, pib.y,  **{'marker': 'o', 'label': 'Nivel'}) # Al objeto axis1 se le aplica un método para dibujar en él la serie.
axis[0].plot(pib.fecha, pib.sm_y,  **{'marker': 'o', 'label': 'Promedio móvil*'})

axis[0].legend()


axis[0].set_ylabel(
    'Millones de Quetzales' , 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)

# Ejes inferiores

axis[1].plot(pib.fecha, pib.v_y,  **{'marker': 'o', 'label': 'Tasa de Nivel**'}) 
axis[1].plot(pib.fecha, pib.v_sm_y,  **{'marker': 'o', 'label': 'Tasa de Promedio móvil**'})

axis[1].legend()


axis[1].set_xlabel(
    'Fecha', 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)

axis[1].set_ylabel(
    'Porcentaje' , 
    **{'fontname':'segoe ui', 'fontsize':'large'}
)



axis[0].set_title(
    r'$\bf{Producto \; Interno \; Bruto \; de \; Guatemala}$' + '\n 2013Q1 a 2019Q4',
    **{'fontname':'segoe ui','fontsize':'x-large'}
)

# Global para toda la figura

txt="Fuente: Banco de Guatemala \n *Promedio móvil de 4 trimestres \n **Tasa de variación iteranual"
plt.figtext(
    0.5, -0.1, txt,
    **{'fontname':'segoe ui','fontsize':'large','wrap':True, 'horizontalalignment': 'center'}
)

plt.savefig('pib3.png')

plt.show()
```
